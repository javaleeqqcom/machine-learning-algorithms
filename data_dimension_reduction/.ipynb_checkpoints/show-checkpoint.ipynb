```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import MDS,Isomap
from sklearn.manifold import LocallyLinearEmbedding as LLE
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from Dim_RD import Dim_RD
from myPCA import myPCA
import os
print(os.getcwd())
```

```python
with open("matteonormb.txt", "r") as f:
    raw_data=f.read()
# print(raw_data)
_data3D=np.array([list(map(float,line.split()[1:])) for line in raw_data.split('\n')])
print(_data3D.shape)
```

```python
data3D=_data3D.copy()
p3D = plt.axes(projection='3d')
xdata,ydata,zdata = tuple(data3D.transpose())
p3D.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Reds')
plt.title("3D")
plt.show()
```

```python
print(data3D.shape)
pca = PCA(copy=True,n_components = 2)
data2D_pca=pca.fit_transform(data3D)
plt.scatter(*tuple(data2D_pca.transpose() ),c= "red")
plt.title("sklearn PCA")
plt.show()
```

```python
print(data3D.shape)
data2D_pca_mine=Dim_RD.pca_fit(data3D,dim_goal=2)
plt.scatter(*tuple(data2D_pca_mine.transpose() ),c= "red")
plt.title("my PCA")
plt.show()
```

```python
mds=MDS(n_components=2)
data2D_mds=mds.fit_transform(data3D)
plt.scatter(*tuple(data2D_mds.transpose() ),c= "blue")
plt.title("sklearn MDS")
plt.show()
```

```python
iso=Isomap(n_components=2)
data2D_mds=iso.fit_transform(data3D)
plt.scatter(*tuple(data2D_mds.transpose() ),c= "blue",marker='.')
plt.title("sklearn Isomap")
plt.show()
```

```python
lle=LLE(n_neighbors =10,n_components=2)
data2D_mds=lle.fit_transform(data3D)
plt.scatter(*tuple(data2D_mds.transpose() ),c= "red",marker='.')
plt.title("sklearn LLE")
plt.show()
```
